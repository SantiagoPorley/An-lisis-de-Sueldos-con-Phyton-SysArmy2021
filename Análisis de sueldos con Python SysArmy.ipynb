7. Printear porcentajes de nivel de educación formal y estado, es decir: % educación secundaria, terciaria, universitaria, postgrado, doctorado, postdoctorado, completo, incompleto y en curso para cada uno.

````python
education_state_list = []

for case in cases_list:
    if case['nivel educacion'] not in education_state_list and case['nivel educacion']  != "":
        education_state_list.append(case['nivel educacion'])
````

````python
#
education_state_dict = dict.fromkeys(education_state_list)
````

````python
#

for level in education_state_list: 
    
    level_state = 0
    
    for case in cases_list:       
        if case["nivel educacion"] == level:
            level_state += 1            
            
    education_state_dict.update({level: {"cantidad": level_state,
                                        "participacion": "",
                                        "incompleto": "",
                                        "cursando": "",
                                        "completado":""}})    
    
````

````python
for i, j  in education_state_dict.items():
    participation = 0    
    participation = round((j["cantidad"] / len(cases_list) * 100) , 2)   
    
    education_state_dict[i].update({"participacion": participation })    
````

````python
for i, j in education_state_dict.items():       
    completado = 0
    cursando = 0
    incompleto = 0
    
    for case in cases_list:        
        if case['nivel educacion'] == i: 
            
            if case["estado nivel"] == "Completado":                
                completado += 1  
                
            if case["estado nivel"] == "En curso":
                cursando += 1
                
            if case["estado nivel"] == "Incompleto":                
                incompleto += 1  
        
    education_state_dict[i].update({"completado": completado, 
                                    "cursando": cursando,
                                     "incompleto": incompleto})    
````

````python
for key , value in education_state_dict.items():
    print("Nivel " + key + ": \n" +  "Porcentaje de participación: " + str(value['participacion']) + "%\n De los cuales "+ str(round((value['completado'] / value['cantidad'] * 100),2)) + "% completaron el nivel, " + str(round((value['cursando'] / value['cantidad'] * 100),2)) + "% cursan el nivel y "+ str(round((value['incompleto'] / value['cantidad'] * 100),2)) + "% no completaron el nivel.\n"  )
````
